### Import Modules

```python
import numpy as np
import pandas as pd
# import geopandas as gpd
# import fiona
import folium
import shapefile
from folium.plugins import MarkerCluster
from folium.features import ColorLine
from folium.features import GeoJson
```

```python
Seattle_coords = [47.6062, -122.3321]
```

```python
fp_shape = '../data/Transit_Routes_for_King_County_Metro__transitroute_line.shp'
sf = shapefile.Reader(fp_shape)
sf_records = sf.shapeRecords() # Simultaneous reading of shape and records
sf_test = sf_records[140] #Route_45 - 140, 579
# Convert to Geojson format
sf_data = sf_test.__geo_interface__
# sf_data
```

```python
route_45_map = folium.Map(location = Seattle_coords, zoom_start = 12)
route_45_map.add_child(GeoJson(sf_data))
route_45_map.fit_bounds(route_45_map.get_bounds())
route_45_map
```
