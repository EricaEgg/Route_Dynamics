```python
# Add module path to list python checks for imports
import os 
import sys
module_path = os.path.abspath(os.path.join('..'))
sys.path.append(module_path)
```

```python
# Import new module
from route_dynamics.route_energy import longi_dynam_model as ldm
```

```python
# Still loads files on framework from last quarters Route_Dynamics 
shapefile_name = '../data/six_routes.shp'
rasterfile_name = '../data/seattle_dtm.tif'
route_num = 45
# avalible routes: [48, 50, 75, 7, 45, 40]
```

```python
some_points_for_stops = [
    (-122.30439282260569, 47.65007615230654),
    (-122.30458069194579, 47.64938740558821)
    ]
```

```python
mass_per_stop_route_instance = ldm.RouteTrajectory(
    route_num, 
    shapefile_name, 
    rasterfile_name,
    bus_speed_model='const_accel_between_stops_and_speed_lim',
    stop_coords=some_points_for_stops,
    mass_array=[22927, 32927]
)
```

```python
# Check out the mass column changing at the stops (at points 5 and 8)
mass_per_stop_route_instance.route_df.head(10)
```

```python
# also check out what happens at the end, mass goes back tot he loaded bus mass
mass_per_stop_route_instance.route_df.tail()
```

```python
# If mass array is not equal length to stop array 
mass_per_stop_route_instance = ldm.RouteTrajectory(
    route_num, 
    shapefile_name, 
    rasterfile_name,
    bus_speed_model='const_accel_between_stops_and_speed_lim',
    stop_coords=some_points_for_stops,
    mass_array=[22927, 32927, 50000000]
)
```

```python
# If mass array includes values that are smaller than the unloaded bus mass
mass_per_stop_route_instance = ldm.RouteTrajectory(
    route_num, 
    shapefile_name, 
    rasterfile_name,
    bus_speed_model='const_accel_between_stops_and_speed_lim',
    stop_coords=some_points_for_stops,
    mass_array=[27, 32927]
)
```

```python
# The unloaded bus mass can be changed as an argument
# If mass array includes values that are smaller than the unloaded bus mass
mass_per_stop_route_instance = ldm.RouteTrajectory(
    route_num, 
    shapefile_name, 
    rasterfile_name,
    bus_speed_model='const_accel_between_stops_and_speed_lim',
    stop_coords=some_points_for_stops,
    mass_array=[27, 32927],
    unloaded_bus_mass=1
)
```
