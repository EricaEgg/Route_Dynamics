## Manually setting bus stops

```python
# First, lets look at the route coordinates for route 45 and pick some for bus stops.
some_points_for_stops = [
    route_instance.route_df.coordinates.iloc[5],
    route_instance.route_df.coordinates.iloc[9]
    ]

some_points_for_stops
    
    
```

```python
# Build new route instance with bus stops marked
stop_example_route_instance = ldm.RouteTrajectory(
    route_num, 
    shapefile_name, 
    rasterfile_name,
    stop_coords=some_points_for_stops)
```

```python
# Check out the top of new DataFrame with bus stops marked 
# and velocities set to zero at those points. 
stop_example_route_instance.route_df.head(10)
```

There seems to be some unit issues with the acceleration, probably because the distances are calculated from coordinates, which are not in Meters.  

```python
# Check that adding bus stops increases power output
energy_1_stop = stop_example_route_instance.energy_from_route() 
energy_no_stop = route_instance.energy_from_route()

print(energy_1_stop)
print(energy_no_stop)
```

## Realistic speed model

```python
import matplotlib.pyplot as plt
def diog_plot(inst, style='plot'):
    
    fig, axes = plt.subplots(3,2,dpi=100, figsize=(8,6))
    x = inst.route_df.cum_distance.values
    v = inst.route_df.velocity.values
    a = inst.route_df.acceleration.values
    t = inst.route_time

    axes[0,0].plot(t,a)
    axes[0,0].set_xlabel('time')
    axes[0,0].set_ylabel('acceleration')


    axes[1,0].plot(t,v)
    axes[1,0].set_xlabel('time')
    axes[1,0].set_ylabel('speed')

    axes[2,0].plot(t,x)
    axes[2,0].set_xlabel('time')
    axes[2,0].set_ylabel('distance')


    axes[0,1].plot(x,a)
    axes[0,1].set_xlabel('distance')
    axes[0,1].set_ylabel('acceleration')

    axes[1,1].plot(x,v)
    axes[1,1].set_xlabel('distance')
    axes[1,1].set_ylabel('speed')

    axes[2,1].plot(x,t)
    axes[2,1].set_ylabel('time')
    axes[2,1].set_xlabel('distance')

    fig.tight_layout()
```

```python
stop_example_route_instance = ldm.RouteTrajectory(
    route_num, 
    shapefile_name, 
    rasterfile_name,
    bus_speed_model='const_accel_between_stops_and_speed_lim',
    stop_coords='random')
```

```python
diog_plot(stop_example_route_instance)
```
